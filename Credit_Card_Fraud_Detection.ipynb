```python
#import module and retriving data
import pandas as pd
cc=pd.read_csv('https://www.dropbox.com/s/6qcgvoc6h8y8zb2/CreditCardDefault.csv?dl=1')
#read the data
cc.head()
```

```python
# statistics
cc.describe()
```

```python
# see the columns involved in the dataset
cc.columns
```

```python
# define X and y
y=cc['Class']
X=cc[['Time', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10',
       'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20',
       'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'Amount']]
```

```python
#dataset split into train and test model
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.7,random_state=2369)
```

```python
#select the ML model
from sklearn.tree import DecisionTreeClassifier
DTC=DecisionTreeClassifier()
#fitting
DTC.fit(X_train,y_train)
```

```python
#prediction 
y_pred=DTC.predict(X_test)
```

```python
#check accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
